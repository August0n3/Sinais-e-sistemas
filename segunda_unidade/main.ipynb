```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do filtro
R = 10e3       # 10 kΩ
C = 10e-9      # 10 nF

# Vetor de frequências (1 Hz a 1 MHz)
f = np.logspace(0, 6, 1000)
w = 2 * np.pi * f

# Resposta em magnitude em dB
H_dB = -10 * np.log10(1 + (w * R * C)**2)

# Frequência de corte
fc = 1 / (2 * np.pi * R * C)  # ≈ 1591,55 Hz

# Plot
fig, ax = plt.subplots(figsize=(9, 5))
ax.semilogx(f, H_dB, linewidth=2, label='|H(ω)| (dB)')
ax.axhline(-3, linestyle='--', label='-3 dB', color='#F97316')
ax.axvline(fc, linestyle='--', label=f'fc = {fc:.1f} Hz')
ax.set_xlabel('Frequência (Hz)')
ax.set_ylabel('Magnitude (dB)')
ax.set_title('Resposta em Magnitude – Filtro Passa-Baixas RC')
ax.legend()
ax.grid(True)
plt.show()
```
